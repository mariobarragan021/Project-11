```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats as st
import math as mth
import datetime as dt
```

```python
hypothesis=pd.read_csv('/datasets/hypotheses_us.csv',sep=";")
hypothesis.info()
```

```python
hypothesis.head(10)
```

```python
#Se crea la columna ICE para reflejar la multiplicacion de los factores,impacto, confianza y esfuerzo

hypothesis['ICE']=((hypothesis['Impact']*hypothesis['Confidence'])/hypothesis['Effort']).round(2)

hypothesis=hypothesis.sort_values(by='ICE',ascending=False)

hypothesis
```

# Conclusion
Como podemos observar la hypothesis que presenta mayor ratio de ICE es la hypothesis 8 ya que tiene los valores mas altos de impacto y confianza y un numero relativamente menor en la cuestion de esfuerzo. 

```python
# calculo para la metrica de RICE

hypothesis['RICE']=(hypothesis['Reach']*hypothesis['Impact']*hypothesis['Confidence'])/hypothesis['Effort'].round(2)

hypothesis.head(9)
```

```python
hypothesis.sort_values(by='RICE',ascending=False,inplace=True)

hypothesis.head(9)
```

# Conclusion
Como podemos observar al realizar el calculo del metrico RICE que tiene el alcance de cada una de las diferentes hypothesis, podemos notar que la hypothesis que tiene mayor alcance es la hypothesis 7 con esto es la primera en la prioridad para realizar. 

```python
hypothesis.groupby('Hypothesis').agg({"ICE":'sum','RICE':'sum'}).sort_values(by='RICE',ascending=False)
```

# Conclusion Primera Fase
Como vemos en la tabla de arriba, las hypothesis cambian con RICE la que tiene la prioridad 1 es Add a subscription form to all the main pages. This will help you compile a mailing list pero si lo vemos con ICE la mejor es Launch a promotion that gives users discounts on their birthdays esto cambia debido al alcance que se tiene en cada una de ellas.

Para la primera el alcance es de 10, mientras que para la segunda el alcance es de 1. 

Se escogeria la que tiene el metodo RICE ya que tiene mayor alcance.

# Inicio Segunda Fase Prueba de hypothesis A/B

```python
#Hypothsis nula: se agrego una forma de suscripcion que ayuda a la lista de los correos
#Hypoteshis alternativa: no se agrega y no tiene un impacto significativo en los usuarios.

ordenes=pd.read_csv('/datasets/orders_us.csv')
ordenes.info()
```

```python
ordenes['date']=ordenes['date'].map(lambda x:dt.datetime.strptime(x,'%Y-%m-%d'))
```

```python
ordenes.info()
```

```python
visitantes=pd.read_csv('/datasets/visits_us.csv')
visitantes.info()
```

```python
visitantes['date']=visitantes['date'].map(lambda x:dt.datetime.strptime(x,'%Y-%m-%d'))
```

```python
visitantes.info()
```

```python
#separar por grupos y quedarnos con valores unicos
ordenes_sin_duplicados=ordenes[['date','group']].drop_duplicates()
```

```python
ordenes_sin_duplicados
```

```python
#se agrgupan los valores sin duplicados
ordenesaggregated=ordenes_sin_duplicados.apply(lambda x: ordenes[np.logical_and(ordenes['date']==x['date'],ordenes['group']==x['group'])]
                                               .agg({
                                                   'date':'max',
                                                   'group':'max',
                                                   'transactionId':pd.Series.nunique,
                                                   'visitorId':pd.Series.nunique,
                                                   'revenue':'sum'
                                               }),axis=1
                                              
                                              ).sort_values(by=['date','group'])
```

```python
ordenesaggregated.columns=['date','group','numeros_pedidos','numero_visitantes','total_ventas']
```

```python
ordenesaggregated
```

```python
visitantes_sin_duplicados=visitantes[['date','group']].drop_duplicates()
```

```python
visitantesaggregated=visitantes_sin_duplicados.apply(lambda x: visitantes[np.logical_and(visitantes['date']==x['date'],visitantes['group']==x['group'])]
                                               .agg({
                                                   'date':'max',
                                                   'group':'max',
                                                   'visits':'sum'
                                               }),axis=1
                                              
                                              ).sort_values(by=['date','group'])
```

```python
visitantesaggregated.columns=['date','group','numero_visitas']
```

```python
visitantesaggregated
```

# Conclusion
Primeramente se limpian los datos quitando duplicados por los diferentes grupos y los dias, posteriormente se agrupan tanto en el dataframe de visitas como el de ordenes para tener una mejor vision de lo que sucede con los diferentes grupos y las hypothesis planteadas. 
Hypothsis nula: se agrego una forma de suscripcion que ayuda a la lista de los correos
Hypoteshis alternativa: no se agrega y no tiene un impacto significativo en los usuarios.

```python
cumulativedata=ordenesaggregated.merge(visitantesaggregated,left_on=['date','group'],right_on=['date','group'])
```

```python
cumulativedata
```

# Conclusion
Se agrupan los datos tanto de las ordenes como de las visitas en un solo dataframe que nso ayudara a sacar metricas importantes mas adelante.

```python
cumulativerevenueA=cumulativedata[cumulativedata['group']=='A'][['date','total_ventas','numeros_pedidos']]
cumulativerevenueB=cumulativedata[cumulativedata['group']=='B'][['date','total_ventas','numeros_pedidos']]
```

```python
plt.plot(cumulativerevenueA['date'],cumulativerevenueA['total_ventas'],label='A')
plt.plot(cumulativerevenueB['date'],cumulativerevenueB['total_ventas'],label='B')

plt.legend()
plt.xticks(rotation=45)
plt.axis([pd.to_datetime('2019-08-01'),pd.to_datetime('2019-08-31'),0,25000])
plt.title('Total de ventas acumuladas por dia y por Grupo')
plt.show()
```

# Conclusion Grafica
Podemos observar que con el paso del tiempo los grupos se compartan muy similar salvo en la fecha muy cercana al17 de agosto donde se ve un pico en el grupo b que hace sobresalir la grafica, esto nos puede afectar en la hypotehis que tenemos. 

```python
plt.plot(cumulativerevenueA['date'],cumulativerevenueA['total_ventas']/cumulativerevenueA['numeros_pedidos'],label='A')
plt.plot(cumulativerevenueB['date'],cumulativerevenueB['total_ventas']/cumulativerevenueB['numeros_pedidos'],label='B')

plt.legend()
plt.xticks(rotation=45)
plt.title('Tamano promedio de compra por dia por Grupo')
plt.show()
```

# Conclusion Grafica
Al igual que en la grafica anterior de tamano de compra o ventas, la tabla de promedio de compra por grupo muestra esa variacion grande en la misma fecha antes mencionada, los demas valores no parecen fluctuar mucho. 

```python
mergedcumulativerevenue=cumulativerevenueA.merge(cumulativerevenueB,left_on='date',right_on='date',how='left',suffixes=['A','B'])
```

```python
mergedcumulativerevenue
```

```python
plt.plot(mergedcumulativerevenue['date'],(mergedcumulativerevenue['total_ventasB']/mergedcumulativerevenue['numeros_pedidosB'])/(mergedcumulativerevenue['total_ventasA']/mergedcumulativerevenue['numeros_pedidosA'])-1)
plt.axhline(y=0,color='black',linestyle='--')
plt.title('Diferencia Relativa para los promedios de compra')
plt.xticks(rotation=45)
plt.show()
```

# Conclusion Grafica
En la grafica nos muestra la relacion que tiene B que es la hypoteshis con cambio con respecto a la A, en el cual podemos observar que B tiene una significancia y va a la alza con respecto a a en el ultimo tramo.

```python
cumulativedata['conversion']=(cumulativedata['numeros_pedidos']/cumulativedata['numero_visitantes']).round(4)
```

```python
cumulativedataA=cumulativedata[cumulativedata['group']=='A']
cumulativedataB=cumulativedata[cumulativedata['group']=='B']

plt.plot(cumulativedataA['date'],cumulativedataA['conversion'],label='A')
plt.plot(cumulativedataB['date'],cumulativedataB['conversion'],label='B')
plt.legend()
plt.xticks(rotation=45)
plt.title('Tasa de conversion por Grupo')
plt.show()
```

# Conclusion Grafica
Con respecto a la tasa de conversion, ambas van disminuyendo con forme la prueba avanza de manera gradual, esto significa que la prueba se esta estabilizando y ambos grupos tambien lo hacen significativamente.

```python
ordersbyusersA=ordenes[ordenes['group']=='A'].groupby(['visitorId'],as_index=False).agg({'transactionId':pd.Series.nunique})

ordersbyusersA.columns=['visitorId','orderId']

ordersbyusersA
```

```python
ordersbyusersB=ordenes[ordenes['group']=='B'].groupby(['visitorId'],as_index=False).agg({'transactionId':pd.Series.nunique})

ordersbyusersB.columns=['visitorId','orderId']

ordersbyusersB
```

```python
sampleA=pd.concat([ordersbyusersA['orderId'],pd.Series(0,index=np.arange(visitantes[visitantes['group']=='A']['visits'].sum()-len(ordersbyusersA['orderId'])),name='orders')],axis=0)
```

```python
sampleB=pd.concat([ordersbyusersB['orderId'],pd.Series(0,index=np.arange(visitantes[visitantes['group']=='B']['visits'].sum()-len(ordersbyusersB['orderId'])),name='orders')],axis=0)
```

```python
print('{0:.3f}'.format(st.mannwhitneyu(sampleA,sampleB)[1])) #valor de p es menor de 0.05
print('{0:.3f}'.format((sampleB.mean()/sampleA.mean())-1)) #perdida o ganancia relativa de B gana en un 13%
```

Con las muestras con datos abnormales podemos definir que no existe tal diferencia estadistica con un valor de significancia de 0.05 y valor de p de 0.017 

```python
print("{0:.3f}".format(st.mannwhitneyu(ordenes[ordenes['group']=='A']['revenue'],
                                    ordenes[ordenes['group']=='B']['revenue'])[1]))

print("{0:.3f}".format((ordenes[ordenes['group']=='B']['revenue'].mean()/
                      ordenes[ordenes['group']=='A']['revenue'].mean())-1))

#es mucho mayor que 0.05 por lo cual no hay una significancia estadistica que rechacemos hipotehsis nula
#por lo tanto el incremento de B se da en un 25.2% con este nuevo cambio en cuanto a revenue se refiere y con los datos en bruto
```

```python
#usuarios anomalos con mayor de 2 pedidos que es el 95%
#usuarios anomalos con mayor de 436 usd en pedidos que es el 95%

userswithmanyorders=pd.concat([ordersbyusersA[ordersbyusersA['orderId']>2]['visitorId'],ordersbyusersB[ordersbyusersB['orderId']>2]['visitorId']],axis=0)
usersexpensiveorders=ordenes[ordenes['revenue']>436]['visitorId']

abnormalusers=pd.concat([userswithmanyorders,usersexpensiveorders],axis=0).drop_duplicates().sort_values()

print(abnormalusers.head())
print(abnormalusers.shape) #son 73 los usuarios anomalos los que se tienen que eliminar del analisis ya que forman parte del 5%
```

```python
sampleAFiltered = pd.concat([ordersbyusersA[np.logical_not(ordersbyusersA['visitorId'].isin(abnormalusers))]['orderId'],pd.Series(0, index=np.arange(visitantes[visitantes['group']=='A']['visits'].sum() - len(ordersbyusersA['orderId'])),name='orders')],axis=0)
```

```python
sampleBFiltered = pd.concat([ordersbyusersB[np.logical_not(ordersbyusersB['visitorId'].isin(abnormalusers))]['orderId'],pd.Series(0, index=np.arange(visitantes[visitantes['group']=='B']['visits'].sum() - len(ordersbyusersB['orderId'])),name='orders')],axis=0)
```

```python
print("{0:.3f}".format(st.mannwhitneyu(sampleAFiltered,sampleBFiltered)[1]))
print("{0:.3f}".format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```
